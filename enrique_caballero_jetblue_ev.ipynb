```python
# # This program performs a DCF analysis to estimate JetBlue's EV
import pandas as pd 
from IPython.display import display
```

```python
# Creating a simple data structure using the provided data
data = {
    "Year": [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010],
    "NOPAT": [18, 53, 89, 120, 149, 181, 215, 247, 270, 292],  # Given data
    "Capital Expenditure": [234, 290, 328, 345, 310, 326, 342, 299, 157, 132],  # Given data
    "Net Working Capital": [34, 63, 94, 126, 157, 191, 227, 261, 285, 308],  # Given data
}
df = pd.DataFrame(data)
display(df)
```

```python
#change in Net Working Capital
df["Change in NWC"] = df["Net Working Capital"].diff().fillna(0)
#free Cash flow
df["FCF"] = df["NOPAT"] - df["Capital Expenditure"] - df["Change in NWC"]
#
display(df)
```

```python
discount_rate = 0.0997 
forecast_period_value = 658
final_year_fcf = df["FCF"].iloc[-1]
perpetual_growth_rate = 0.03
forecast_years = 10
# Terminal Value using the GGMM
terminal_value = (final_year_fcf * (1 + perpetual_growth_rate)) / (discount_rate - perpetual_growth_rate)
print(f"The calculated Terminal Value (TV) is: ${terminal_value:.2f} million")
```

```python
# TV to PV
pv_of_terminal_value = terminal_value / (1 + discount_rate) ** forecast_years
# EV
enterprise_value = forecast_period_value + pv_of_terminal_value
#
print(f"The Present Value (PV) of Terminal Value is: ${pv_of_terminal_value:.2f} million")
print(f"The Total Enterprise Value (EV) is: ${enterprise_value:.2f} million")
```
